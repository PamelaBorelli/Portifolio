```python
import warnings
warnings.filterwarnings('ignore')

import os
import pandas as pd

# Pasta raiz
pasta_raiz = '/content/CSTNews/'

# Cria dois DataFrames vazios
df = pd.DataFrame(columns=['id', 'texto'])
df_sumarios = pd.DataFrame(columns=['id', 'sumario_humano'])

# Percorre todos os subdiretórios
for subdir, dirs, files in os.walk(pasta_raiz):
    # Verifica se o subdiretório atual é 'Textos-fonte segmentados'
    if os.path.basename(subdir) == 'Textos-fonte segmentados':
        # Lê o conteúdo de cada arquivo e adiciona ao DataFrame
        for arquivo in files:
            caminho_do_arquivo = os.path.join(subdir, arquivo)
            try:
                with open(caminho_do_arquivo, 'r') as f:
                    conteudo = f.read()
                # Remove as extensões do nome do arquivo
                arquivo_sem_extensao = os.path.splitext(os.path.splitext(arquivo)[0])[0]
                df = df.append({'id': arquivo_sem_extensao, 'texto': conteudo}, ignore_index=True)
            except Exception as e:
                print(f'Erro ao abrir o arquivo {caminho_do_arquivo}: {e}')
    # Verifica se o subdiretório atual é 'Sumarios'
    elif os.path.basename(subdir) == 'Sumarios':
        # Lê o conteúdo de cada arquivo que começa com 'D' e termina com 'sumario_humano' e adiciona ao DataFrame
        for arquivo in files:
            if arquivo.startswith('D') and arquivo.endswith('sumario_humano.txt'):
                caminho_do_arquivo = os.path.join(subdir, arquivo)
                try:
                    with open(caminho_do_arquivo, 'r', encoding='utf-8') as f:
                        conteudo = f.read()
                    # Remove as extensões do nome do arquivo
                    arquivo_sem_extensao = os.path.splitext(os.path.splitext(arquivo)[0])[0]
                    df_sumarios = df_sumarios.append({'id': arquivo_sem_extensao, 'sumario_humano': conteudo}, ignore_index=True)
                except UnicodeDecodeError:
                    try:
                        with open(caminho_do_arquivo, 'r', encoding='latin-1') as f:
                            conteudo = f.read()
                        df_sumarios = df_sumarios.append({'id': arquivo_sem_extensao, 'sumario_humano': conteudo}, ignore_index=True)
                    except Exception as e:
                        print(f'Erro ao abrir o arquivo {caminho_do_arquivo} com a codificação latin-1: {e}')
                        continue
                except Exception as e:
                    print(f'Erro ao abrir o arquivo {caminho_do_arquivo} com a codificação utf-8: {e}')
                    continue
```

```python
# Cria uma lista com os IDs para remover
ids_para_remover = ['D1_C30_Folha_07-08-2007_09h19_versao3', 'D1_C30_Folha_07-08-2007_09h19_versao2', 'D1_C30_Folha_07-08-2007_09h19_versao1', 'D3_C39_OGlobo_07-08-2007_13h31_versao2', 'D3_C39_OGlobo_07-08-2007_13h31_versao1', 'D3_C18_GPovo_versao1', 'D2_C7_OGlobo_versao2', 'D2_C7_OGlobo_versao3', 'D2_C7_OGlobo_versao1', 'D4_C50_JB_06-11-2007_14h12_versao1']

# Remove as linhas do DataFrame
df_sumarios = df_sumarios[~df_sumarios['id'].isin(ids_para_remover)]

# Reinicia o índice do DataFrame
df_sumarios = df_sumarios.reset_index(drop=True)

# Imprime o DataFrame
df_sumarios
```

```python
# Encontra as linhas onde a coluna 'id' contém a palavra 'versao'
linhas = df_sumarios[df_sumarios['id'].str.contains('versao')]

# Imprime as linhas
linhas['id']
```

```python
# Use uma expressão regular para substituir tudo a partir de '_versao'
df_sumarios['id'] = df_sumarios['id'].str.replace(r'_versao.*$', '', regex=True)

# Imprime o DataFrame
df_sumarios
```

```python
# Cria uma lista com os índices das linhas para selecionar
linhas_para_selecionar = [22, 48, 98, 100, 108]

# Seleciona as linhas do DataFrame
df_selecionado = df_sumarios.iloc[linhas_para_selecionar]

# Imprime o DataFrame selecionado
df_selecionado
```

```python
# Cria uma lista com as expressões para buscar
expressoes_para_buscar = ['D2_C34_Estadao', 'D3_C34_JB', 'D1_C34_Folha', 'D2_C31_Estadao', 'D1_C31_Folha',
'D1_C17_Folha', 'D2_C17_Estadao', 'D2_C32_Estadao', 'D3_C32_JB', 'D4_C32_GPovo']

# Para cada expressão na lista, substitui tudo que vem depois da expressão por ''
for expressao in expressoes_para_buscar:
    df_sumarios['id'] = df_sumarios['id'].str.replace(r'(?<={})_.*$'.format(expressao), '', regex=True)

# Imprime o DataFrame
df_sumarios
```

```python
# Imprime as colunas do DataFrame df
print(df.columns)

# Remove os espaços em branco antes e depois dos nomes dos IDs no DataFrame df
df['id'] = df['id'].str.strip()

# Imprime as colunas do DataFrame df_sumarios
print(df_sumarios.columns)

# Remove os espaços em branco antes e depois dos nomes dos IDs no DataFrame df_sumarios
df_sumarios['id'] = df_sumarios['id'].str.strip()
```
