```python
from datasets import load_dataset

dataset = load_dataset("PamelaBorelli/XLSum_portuguese", split="train")


def get_training_corpus():
    for i in range(0, len(dataset), 1000):
        yield dataset[i : i + 1000]["text"]

training_corpus = get_training_corpus()
```

```python
!pip install tokenizers==0.13.3
from tokenizers import Tokenizer
from tokenizers import models

tokenizer = Tokenizer(models.Unigram())
```

```python
from tokenizers import (
    decoders,
    models,
    normalizers,
    pre_tokenizers,
    processors,
    trainers,
    Tokenizer,
    Regex,
)


tokenizer.normalizer = normalizers.Sequence(
    [
        normalizers.Replace("``", '"'),
        normalizers.Replace("''", '"'),
        normalizers.NFKD(),
        normalizers.StripAccents(),
        normalizers.Replace(Regex(" {2,}"), " "),
    ]
)
```

```python
tokenizer.pre_tokenizer = pre_tokenizers.Metaspace()
```

```python
tokenizer.pre_tokenizer.pre_tokenize_str('O estudo constata que 5% da população espanhola, em nível nacional, teve algum contato com o vírus, ou seja, pouco mais de 2 milhões de pessoas", declarou o ministro da Saúde, Salvador Illa, ao comentar os primeiros resultados parciais desse estudo iniciado em 27 de abril e que testou mais de 60 mil pessoas. "O resultado que mais chamou atenção foi a grande variabilidade geográfica", ressaltou o ministro. O estudo demonstra que o coronavírus circulou muito mais no centro do país, especialmente na capital, do que em outras regiões. As maiores taxas de contágio foram registradas nas vizinhas Castilla e León, com 14,2% em Soria e Castilla-La Mancha por exemplo. Sem imunidade de rebanho Até agora, o governo espanhol relatou apenas os resultados de uma primeira onda de "testes rápidos" para encontrar anticorpos contra o coronavírus. "Não encontramos diferenças entre homens e mulheres, nem nas faixas etárias", embora "seja verdade que crianças e adolescentes pareçam ter uma prevalência mais baixa", explicou o ministro.')
```

```python
special_tokens = ["<unk>", "<pad>", "</s>"]
trainer = trainers.UnigramTrainer(
    vocab_size=32100, special_tokens=special_tokens, unk_token="<unk>"
)
tokenizer.train_from_iterator(get_training_corpus(), trainer=trainer)
```

```python
unk_token_id = tokenizer.token_to_id("<unk>")
pad_token_id = tokenizer.token_to_id("<pad>")
s_token_id = tokenizer.token_to_id("</s>")
print(unk_token_id, pad_token_id, s_token_id)
```

```python
tokenizer.post_processor = processors.TemplateProcessing(
    single="$A:0 $B:1 <unk>:2 <pad>:3 </s>:4",
    special_tokens=[
        ("<unk>", unk_token_id),
        ("<pad>", pad_token_id),
        ("</s>", s_token_id),
    ],
)
```
