```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Carregar o modelo e o tokenizador
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
```

```python
text = "O cão está feliz com o açaí. E o índio toca seu violão sob a lua saúde notícia maçã promoção pedido até às instâncias pódio"
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)
```

```python
encoded_ids = tokenizer.encode(text, add_special_tokens=False)
print("IDs de token:", encoded_ids)
```

```python
# Decodificar os IDs para texto
texto_decodificado = tokenizer.decode(encoded_ids)
print("Texto decodificado:", texto_decodificado)
```

```python
# Adicionar os tokens desejados ao vocabulário
new_tokens = ["ú", "ã", "í", "õ"]
num_added_toks = tokenizer.add_tokens(new_tokens, special_tokens=True)

# Redimensionar a camada de incorporação
old_num_tokens = model.config.vocab_size
model.resize_token_embeddings(old_num_tokens + num_added_toks)
```

```python
print("Tamanho do vocabulário antes de adicionar os tokens:", tokenizer.vocab_size)
print("Tamanho do vocabulário após adicionar os tokens:", tokenizer.vocab_size)
```
